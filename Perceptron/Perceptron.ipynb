## Perceptron liniowy

Klasyfikacja jest dziedzin bardzo popularn i istnieje wiele rnych algorytmw uczenia maszynowego, ktre su do rozwizywania problemw klasyfikacji. My przyjrzymy si dziaaniu podstawowego algorytmu, jednak jego dobre zrozumienie jest kluczowe do pojcia bardziej zaawansowanych metod. 

Perceptron, bo o nim bdzie mowa w tym notatniku, uywany jest do klasyfikacji binarnej. Posiada due ograniczenie, ktre mwi, e algorytm jest zbieny w skoczonej liczbie krokw, jeli zbir danych jest liniowo separowalny. Co oznacza, e zbir jest liniowo separowalny? e prbki nalece do rnych klas mona oddzieli za pomoc prostej w przypadku dwuwymiarowym, paszczyzny w przypadku trjwymiarowym i hiperpaszczyzny w przypadku $N$-wymiarowym.

Spjrzmy jeszcze raz na ankiet prezentujc podzia na pe ze wzgldu na wzrost i wag.



Czy ten zbir jest liniowo separowalny? Jest, poniewa jestemy w stanie narysowa wiele rnych prostych, ktre oddzielaj kobiety od mczyzn. Zadanie perceptronu wanie na tym polega, aby znale prost (w najprostszym przypadku), ktra oddziela od siebie dwie klasy i na podstawie stworzonego podziau klasyfikowa nowe prbki.


### Budowa sztucznego neuronu

Perceptron zosta stworzony na bazie sztucznego neuronu. Twrcami modelu sztuczego neuronu byli McCulloch i Pitts, ktrzy w roku 1943 zainspirowani dziaaniem biologicznego neuronu, zaproponowali jego budow.




Sztuczny neuron skada si na nastpujcych elementw:

* **Wejcie** - warstwa odpowiadajca za przyjmowanie danych wejciowych. Istnieje tyle pocze do neuronu, ile jest cech w zestawie danych. Przykadowo, jeli chcemy rozpoznawa pe na podstawie wagi, wzrostu, dugoci stopy i dugoci wosw, to mamy 4 cechy opisujce osob, wic bd 4 poczenia w warstwie wejciowej.


* **Wagi** - kade poczenie do neuronu ma skojarzon ze sob wag. Wagi s tym, czym byy wspczynniki regresji w algorytmie regresji. Tutaj przechowywana jest "wiedza" neuronu.


* **Sumator** - element, ktry odpowiada za zsmumowanie wszystkich iloczynw wej i wag. Jest to tak zwane pobudzenie neuronu, czyli waona suma wej. Do sumatora wchodzi jeszcze jedno dodatkowe poczenie. Na jego wejciu zawsze podawana jest warto $1$, a waga skojarzona z tym wejciem nazywana jest biasem. Odpowiada ona za przesunicia prostej separujcej klasy. Gdyby jej nie byo, prosta ta zawsze przechodziaby przez pocztek ukadu wsprzdnych.  


* **Funkcja aktywacji** - pobudzenie neuronu obliczone w sumatorze trafia nastpnie do funkcji aktywacji. To ona decyduje o zachowaniu neuronu, czyli o jego odpowiedzi. Istnieje wiele rnych funkcji aktywacji, my pocztkowo wykorzystamy skokow funkcj aktywacji Heavisidea:

$$ f(s) = \left\{\begin{array}{l}
1, \quad s \geq 0,\\
0, \quad s<0\end{array}\right.$$


* **Wyjcie** - odpowied neuronu, czyli przetworzona przez funkcje aktywacji waona suma wej. W przypadku klasyfikacji binarnej bdzie to warto $0$ lub $1$.

Neuron, ktry bdziemy implementowa, ze wzgldu na metod trenowania i wybran funkcj aktywacji nazywany jest **perceptronem**. Wzr matematyczny opisujcy powysz budow perceptronu jest nastpujcy:

$$h_w(x) = f\left(\sum_{i=0}^{n}{w_ix_i}\right)$$

Zgodnie z notacj zaproponowan przez Prof. Andrew Ng, funkcje, ktre mapuj obserwacje na wynik bdziemy nazywa hipotezami (_ang. hypothesis_) i oznacza jako $h_w(x)$. 
Prosz zauway, e we wzorze zosta uwzgldniony bias, czyli waga $w_0$. Na wejcie $x_0$ zawsze podawana jest warto $1$.


### Nauka perceptronu

Istniej rne metody treningu sztucznego neuronu. Ten, ktry zaimplementujemy wykorzystywa bdzie regu perceptronow. To prosta metoda, ktra suy do aktualizacji wag w neuronie ze skokow funkcj aktywacji. Wszystkie wagi perceptronu aktualizowane s jednoczenie po kadej prbce uczcej. Regua perceptronowa moe zosta zapisana jako:

$$
w_{i}=\begin{cases}
w_i,                    &  \ h_w(x^{(k)}) = y^{(k)},\\
w_i + \alpha x_i^{(k)}, &  \ h_w(x^{(k)}) < y^{(k)},\\
w_i - \alpha x_i^{(k)}, &  \ h_w(x^{(k)}) > y^{(k)}
\end{cases}
$$

Gdzie: 
* $y^{(k)}$ jest oczekiwan odpowiedz perceptronu (klas dla $k$-tej prbki uczcej)
* $\alpha$ jest wspczynnikiem uczenia i przybiera wartoci z przedziau $(0,1)$.

Pamitajmy, e $h_w(x)$ i $y$ mog przyj tylko wartoci $0$ lub $1$. W takim przypadku wszystkie moliwe kombinacje aktualizacji wag wygldaj nastpujco:

| h_w(x) | y |     aktualizacja wag     |
|:------:|:-:|:------------------------:|
|    0   | 0 |         bez zmian        |
|    0   | 1 | $w_i + \alpha x_i^{(k)}$ |
|    1   | 0 | $w_i - \alpha x_i^{(k)}$ |
|    1   | 1 |         bez zmian        |

Czyli jeli odpowied perceptronu bya $0$, a rzeczywista warto wyniosa $1$, to musimy zwikszy wagi, aby pobudzenie neuronu miao wiksz warto. Jeli odpowied perceptronu bya $1$, a rzeczywista warto $0$, to musimy zmniejszy wagi, by pobudzenie neuronu miao mniejsz warto. W pozostaych przypadkach odpowiedzi si zgadzaj, wic nie aktualizujemy wag. Dziki temu prostemu algorytmowi, perceptron zawsze znajdzie prost separujc klasy w liniowo separowalnym zestawie danych.


### Algorytm perceptronu

Majc ju wszystkie potrzebne informacje o dziaaniu perceptronu, zapiszmy algorytm.

1. Wygenerowanie pocztkowych wag perceptronu,
2. Wybranie $k$ - tej prbki uczcej ze zbioru danych
3. Obliczenie waonej sumy wej dla $k$-tej prbki uczcej,
4. Oblicznie odpowiedzi neuronu poprzez funkcj aktywacji Heavisidea,
5. Aktualizacja wag zgodnie z regu perceptronow,
6. Powrt do kroku 2 jeli k < N, a N to liczba wszystkich prbek uczcych,
7. Przerwanie uczenia, jeli perceptron rozpoznaje wszystkie prbki prawidowo lub jego bd mieci si w ustalonym zakresie. W przeciwnym wypadku, ustawienie na k = 0 i powrt do kroku 2.

Przejcie przez wszystkie prbki uczce w zestawie danych nazywane jest epok. Punkt 7 mwi o tym, e jeli wszystkie prbki w epoce zostay dobrze sklasyfikowane, to moemy przerwa uczenie. Ilo epok potrzebna do treningu perceptronu jest zalena od wspczynnika uczenia $\alpha$ oraz od zestawu danych. W zestawach danych, ktre nie s liniowo separowalne, perceptron nigdy nie bdzie mia stuprocentowej skutecznoci klasyfikacji. W takich przypadkach dobrze jest zabezpieczy algorytm przed nieskoczonym procesem uczenia.


### Granica decyzyjna

Wiadomo ju, e zadaniem perceptronu jest oddzielenie prbek z przeciwnych klas. Prosta, ktra oddziela prbki z rnych klas nazywana jest granic decyzyjn. Przed przejciem do implementacji, warto wyrobi sobie pewn intuicj, co to tak naprawd to oznacza i dlaczego perceptron w ogle dziaa.

Zamy, e mamy zbir danych z dwoma cechami $x_1$ i $x_2$. Zbir ten jest liniowo separowalny. 


    

W naszym przypadku wzr perceptronu dla dwch cech wyglda nastpujco: 

$$h_w(x) = f(w_0x_0 + w_1x_1 + w_2x_2)$$

Spjrzmy jeszcze jak wyglda wykres naszej funkcji aktywacji Heavisidea, ktra decyduje o odpowiedzi perceptronu.

<div>
    
</div>

Analizujc wykres, mona zauway, e odpowied $0$ bdzie w sytuacji, gdy warto wchodzca do funkcji aktywacji bdzie ujemna, a gdy bdzie dodatnia, to odpowied bdzie rwna $1$. Dobrze, czyli wiemy, e eby perceptron da predykcj $1$, to waona suma wej musi by wiksza od zera, w przeciwnym przypadku da odpowied $0$. Zapiszmy to.

$h_w(x) = 1$ jeli

$$w_0x_0 + w_1x_1 + w_2x_2 > 0$$ 

$h_w(x) = 0$ jeli 

$$w_0x_0 + w_1x_1 + w_2x_2 < 0$$ 

Czyli wynika z tego, e zmiana decyzji predykcji jest wtedy, gdy waona suma bdzie rwna $0$. Wykres funkcji aktywacji to potwierdza.

$$w_0x_0 + w_1x_1 + w_2x_2 = 0$$


Zamy, e za pomoc algorytmu perceptronu znalelimy nastpujce wartoci wag: $[-5, 1, 1]$. Podstawmy te wartoci do rwnania.

$$-5 + 1x_1 + 1x_2 = 0$$ 

Pamitamy, e $x_0$ ma zawsze warto $1$, wic pomijamy je w rwnaniu. Teraz przeksztamy rwnanie tak, aby po lewej stronie byy niewiadome.

$$x_1 + x_2 = 5$$ 
$$x_2 = -x_1 + 5$$

To co otrzymalimy, to rwnanie prostej o wspczynniku kierunkowym $-1$. Prosta ta przetnie o $x_2$ w punkcie $(0,5)$, a o $x_1$ w punkcie $(5,0)$. Naniemy prost na nasz wykres z zestawem danych.



Prosta, ktr nanielimy na wykres, to **granica decyzyjna** (ang. decision boundary). Jest to miejsce miejsce, w ktrym nastpuje zmiana predykcji. Wszystkie prbki, ktre znajd si poniej prostej bd zaklasyfikowane jako klasa $0$, a te powyej, jako klasa $1$.


### Jak narysowa granic decyzyjn

Rwnanie oglne prostej ma posta $Ax + By + C = 0$

W naszym przypadku: 

$w_{1}x_{1} + w_{2}x_{2} + w_{0}x_{0} = 0$, gdzie $x_{0} = 1$

Dlaczego = $0$? Poniewa w tym miejscu nastpuje zmiana decyzji klasyfikacji - zgodnie z wykresem funkcji skokowej Heaviside'a.

Teraz chcemy przeksztaci rwnanie w taki sposb, eby tylko $x_{2}$ byo po lewej stronie stronie rwnania, czyli chcemy przej z rwnania oglnego prostej na rwnanie kierunkowe. Dokadnie tak jak si to robio w szkole redniej np. [przykad](https://zadaniacke.pl/teoria/rownanie-kierunkowe-i-ogolne-prostej/)

$w_{1}x_{1} + w_{2}x_{2} + w_{0}x_{0} = 0$ <- rwnanie OGLNE prostej

$w_{2}x_{2} = -(w_{1}x_{1} + w_{0})$


$x_{2} = \frac{-(w_{1}x_{1} + w_{0})}{w_{2}}$ <- rwnanie KIERUNKOWE prostej

Majc rwnanie kierunkowe prostej w atwy sposb moemy stworzy rysunek granicy decyzyjnej.

Przykadowy kod tworzcy granic decyzyjn w zadanym przedziale wartoci:

```python
import matplotlib.pyplot as plt
import numpy as np

# ustalamy wagi
w = [2, 3, 1.5]

# tworzymy punkty na osi x1
x1 = np.linspace(-5, 5, 100)

# obliczamy x2 (y) dla wszystkich punktw x1 na podstawie wybranych wag
x2 = -(w[1]*x1 + w[0])/w[2]

# podpisujemy osie
plt.xlabel("x1")
plt.ylabel("x2")

# dodajemy dodatkowo osie x i y, eby wiedzie gdzie jest punkt 0,0
zeros = np.zeros(100)
plt.plot(x1, zeros, c="black", linewidth=0.5, linestyle='--')
plt.plot(zeros, x1, c="black", linewidth=0.5, linestyle='--')

# ograniczamy widoczno wykresu do zakresu -5, 5 dla obu osi
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.plot(x1, x2)
plt.show()
```

````python
import unittest

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from typing import Tuple
import time
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
````

````python
df = pd.read_csv('datasets/Ankieta.csv')
df['Plec'] = df['Plec'].map({'Kobieta': 1, 'Mezczyzna': 0})

df
````

````python
def perceptron(X: np.ndarray, y: np.ndarray, alpha: float, epochs_number: int) -> list:
    weights = np.random.uniform(0, 1, len(X[0]) + 1) 
    weights_history = [weights]  
    for _ in range(epochs_number):
        index = 0
        for input, val in zip(X, y):
            input = np.insert(input, 0, 1)
            weighted_sum = input.dot(weights)
            activation = 1 if weighted_sum >= 0 else 0
            
            if activation == val:
                index += 1
            elif activation < val:
                weights = weights + alpha * input
            elif activation > val:
                weights = weights - alpha * input
            
            if index >= 0.99 * len(X):
                weights_history.append(weights.tolist())
                return weights_history
        weights_history.append(weights.tolist())
    return weights_history

````

````python
df = pd.read_csv('datasets/Ankieta.csv', sep=',')
df['Plec'] = df['Plec'].map({'Kobieta': 1, 'Mezczyzna': 0})

start_time = time.time()
weights_history = perceptron(df[['Waga', 'Wzrost']].values, df['Plec'], 0.01, 1000)
end_time = time.time()
original_time = end_time - start_time

normalized_weight = (df['Waga'] - np.mean(df['Waga'])) / np.std(df['Waga'])
normalized_height = (df['Wzrost'] - np.mean(df['Wzrost'])) / np.std(df['Wzrost'])

start_time = time.time()
weights_history = perceptron(pd.DataFrame({'Waga': normalized_weight, 'Wzrost': normalized_height}).values, df['Plec'], 0.01, 1000)
end_time = time.time()
normalized_time = end_time - start_time

print(f"Czas dziaania dla danych nieznormalizowanych: {original_time}")
print(f"Czas dziaania dla danych  znormalizowanych: {normalized_time}")
````

Oblicz skuteczno predyckji perceptronu

````python
X_train, X_test, y_train, y_test = train_test_split(df[['Waga', 'Wzrost']], df['Plec'], test_size=0.2, random_state=42)

X_train_norm = (X_train - X_train.mean()) / X_train.std()
X_test_norm = (X_test - X_train.mean()) / X_train.std()

weights_history = perceptron(X_train_norm.values, y_train, 0.01, 1000)
final_weights = weights_history[-1]


def predict_perceptron(sample):
    weighted_sum = np.dot(sample, final_weights[1:]) + final_weights[0]
    return 1 if weighted_sum >= 0 else 0

y_pred = np.array([predict_perceptron(sample) for sample in X_test_norm.values])

accuracy = accuracy_score(y_test, y_pred)
print(f"Skuteczno predykcji perceptronu: {accuracy * 100}%")
````

Zaproponuj test jednostkowy sprawdzajcy zaimplementowany algorytm

````python
class PerceptronUnitTest(unittest.TestCase):
    def test_perceptron_classification(self):
        X = np.array([[1, 1], [1, 2], [2, 1], [2, 2]])
        y = np.array([0, 0, 1, 1])  

        alpha = 0.01
        epochs_number = 1000

        weights_history = perceptron(X, y, alpha, epochs_number)

        final_weights = weights_history[-1]
        for sample, true_label in zip(X, y):
            sample_with_bias = np.insert(sample, 0, 1)
            prediction = 1 if sample_with_bias.dot(final_weights) >= 0 else 0
            self.assertEqual(prediction, true_label)
        
        
unittest.main(argv=[''], verbosity=2, exit=False)
````
